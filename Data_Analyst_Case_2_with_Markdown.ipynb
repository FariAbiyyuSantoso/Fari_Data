```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Memuat data dari file csv ke dalam DataFrame Pandas untuk analisis
house_categorical = pd.read_csv('house_categorical.csv',index_col=0) #Index_col=0 untuk menjaga keteraturan dan kemudahan akses terhadap data, dengan kolom pertama digunakan sebagai indeks
house_numeric = pd.read_csv('house_numeric.csv',index_col=0)
```

```python
#Visualisasi distribusi kategori dalam kolom Utilities dari DataFrame house_categorical
sns.barplot(x=house_categorical['Utilities'].value_counts().index, y=house_categorical['Utilities'].value_counts(normalize = True))
plt.xlabel('x')            # Menambahkan label sumbu X
plt.ylabel('y')            # Menambahkan label sumbu Y
```

```python
##Visualisasi distribusi kategori dalam kolom HouseStyle dari DataFrame house_categorical
sns.barplot(x=house_categorical['HouseStyle'].value_counts().index, y=house_categorical['HouseStyle'].value_counts(normalize = True))
plt.xlabel('x')            # Menambahkan label sumbu X
plt.ylabel('y')            # Menambahkan label sumbu Y
```

```python
#Visualisasi distribusi kategori dalam kolom ExterQual dari DataFrame house_categorical
sns.barplot(x=house_categorical['ExterQual'].value_counts().index, y=house_categorical['ExterQual'].value_counts(normalize = True))
plt.xlabel('x')            # Menambahkan label sumbu X
plt.ylabel('y')            # Menambahkan label sumbu Y
```

```python
x = house_numeric["SalePrice"] #mengambil kolom SalePrice dari DataFrame house_numeric dan menyimpannya dalam variabel x
sns.distplot(x, axlabel=x.name) #membuat histogram dan kurva distribusi (kernel density estimate) dari data di x
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean") #visualisasi dari nilai rata-rata dalam distribusi & menggambar garis vertikal pada nilai rata-rata dari x (harga jual) dengan garis putus-putus hitam ("k").
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median") #visualisasi dari nilai median dalam distribusi & menggambar garis vertikal pada nilai median dari x dengan garis putus-putus merah ("r").
first_legend = plt.legend(loc=1) #loc=1 menentukan lokasi legenda di sudut kanan atas plot.(Untuk menjelaskan garis rata rata & median)
plt.xlabel('x')            # Menambahkan label sumbu X
plt.ylabel('y')            # Menambahkan label sumbu Y
print("median:", x.median())
print("mean:", x.mean())
print("Std Dev1", x.std())
plt.show()
```

```python
x = np.log10(house_numeric["SalePrice"]) #Menghitung logaritma basis 10 dari argumen yang diberikan, Transformasi ini membantu dalam membuat distribusi data menjadi lebih seimbang
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(loc=1)
plt.xlabel('x')            # Menambahkan label sumbu X
plt.ylabel('y')            # Menambahkan label sumbu Y
print("median:", x.median())
print("mean:", x.mean())
print("Std Dev1", x.std())
plt.show()
```

```python
# Filter data: hanya menyertakan baris di mana GarageArea > 0
filtered_house = house[house['GarageArea'] > 0]

# Tampilkan hasil filter
print(filtered_house)
```

```python
#Analisis hubungan antara dua variabel numerik: luas area garasi (GarageArea yang sudah terfilter) dan harga jual rumah(SalePrice)
ax = sns.scatterplot(x="GarageArea", y="SalePrice", data=filtered_house)
plt.title("Linear graph")
plt.xlabel('x')            # Menambahkan label sumbu X
plt.ylabel('y')            # Menambahkan label sumbu Y
plt.show()
```

```python
# Menghitung korelasi antara SalePrice dan GrLivArea
corr_saleprice_grlivarea = house['SalePrice'].corr(house['GrLivArea'])

# Menghitung korelasi antara SalePrice dan GarageArea
corr_saleprice_garagearea = house['SalePrice'].corr(house['GarageArea'])

# Menampilkan hasil korelasi
print(f"Korelasi SalePrice dengan GrLivArea: {corr_saleprice_grlivarea}")
print(f"Korelasi SalePrice dengan GarageArea: {corr_saleprice_garagearea}")
```
